### 12. Replace missing postal codes with 0

```python
sales_forecasting_data_updated_missing_postal_codes = sales_forecasting_data["Postal Code"].fillna(0)

missng_value = sales_forecasting_data_updated_missing_postal_codes.isnull().sum()

print("There are missing 'Postal Code' in the dataset." if bool(missng_value > 0) else "There is no missing 'Postal Code' in the dataset." )
```

### 13. Convert the "Order Date" column to datetime type

```python
print("Default Order Date Column dtype: ",sales_forecasting_data["Order Date"].dtype)
sales_forecasting_data["Order Date"]=pd.to_datetime(sales_forecasting_data["Order Date"], dayfirst=True)

print("Updated Order Date Column dtype: ",sales_forecasting_data["Order Date"].dtype)
```

### 14. Extract the year, month and day from 'Order Date'

```python
sales_forecasting_data_order_dates = sales_forecasting_data.head(10)["Order Date"]

days = sales_forecasting_data_order_dates.dt.day


print("Order Days: ")
print(days)


```

```python
months = sales_forecasting_data_order_dates.dt.month

print("Order Months: ")
print(months)
```

```python
years = sales_forecasting_data_order_dates.dt.year.head(10)

print("Order Years: ")
print(years)
```

### 15. Create a new column Order_YearMonth combining year and month

```python
sales_forecasting_data["Order_YearMonth"] = sales_forecasting_data["Order Date"].dt.year.astype(str) + "-" + sales_forecasting_data["Order Date"].dt.month.astype(str)
print("Updated Dataset: ")
sales_forecasting_data.head(5)
```

### 16. Rename the column "Postal Code" to "Postal_Code"

```python
sales_forecasting_data.rename(columns={"Postal Code": "Postal_Code"},inplace=True)

sales_forecasting_data.head(5)
```

### 17. Set "Order ID" as index

```python
sales_forecasting_data.rename(columns={"Order ID": "index"}, inplace=True)
sales_forecasting_data.head(3)
```
