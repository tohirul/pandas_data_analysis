### 14.1: filtering based on string

```python
data = {
    "Name":["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah"],
    "City": ["New York", "Los Angeles", "Newark", "Boston", "New Delhi", "Chicago", "New Orleans", "Houston"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Marketing", "Finance", "HR"],
    "Salary":[50000, 60000, 55000, 70000, 52000, 58000, 62000, 51000]
}

employee_profile = pd.DataFrame(data)
```

```python
employee_profile.index+=1

employee_profile
```

```python
### starts with new 

loc_condition = employee_profile['City'].str.contains('new', case=False)

print(loc_condition)
employee_profile.loc[loc_condition]
```

```python
### starts with Los

loc_condition = employee_profile['City'].str.contains(r'^Los')

print(loc_condition)
employee_profile.loc[loc_condition]
```

```python
### ends with rk

employee_profile.loc[employee_profile['City'].str.contains(r'rk$')]
```

```python
### name starts with vowel

employee_profile.loc[employee_profile['Name'].str.contains(r'^[AEIOU]')]
```

```python
# string contains new or los
employee_profile.loc[employee_profile['City'].str.contains(r'New|Los')]
```

### 14.7: Apply Function in DF

```python
# min max scaling
```

```python
mn = student_data['Total Marks'].min()

mx = student_data['Total Marks'].max()
```

```python
student_data['Scaled Marks'] = student_data['Total Marks'].apply(lambda x: (x-mn)/(mx-mn))
```

```python
student_data
```

```python
def SET_Grade(total_marks):
    if total_marks >= 270:
        return 'A+'
    elif total_marks >=260:
        return 'A'
    else:
        return 'F'
        
```

```python
student_data['GPA'] = student_data['Total Marks'].apply(SET_Grade)
```
