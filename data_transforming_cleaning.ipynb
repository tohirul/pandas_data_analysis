```python
import pandas as pd
```

```python
sales_forecasting_data = pd.read_csv('sales-forecasting/train.csv')

sales_forecasting_data
```

### 1. Add a new column "Profit Margin" = Sales*0.1 (simulate)

```python
sales_forecasting_data['Profit Margin'] = (sales_forecasting_data['Sales']*0.1).round(4)

sales_forecasting_data
```

### 2. Create a column 'Shipping Days'= 'Ship Date - Order Date'

```python
sales_forecasting_data['Shipping Days'] = pd.to_datetime(sales_forecasting_data['Ship Date'],dayfirst=True) - pd.to_datetime(sales_forecasting_data['Order Date'],dayfirst=True)

sales_forecasting_data
```

### 3. Convert 'Shipping Days' to integer days

```python
sales_forecasting_data["Shipping Days"] = (
    pd.to_datetime(sales_forecasting_data['Ship Date'],dayfirst=True) - pd.to_datetime(sales_forecasting_data['Order Date'],dayfirst=True)
).dt.days

sales_forecasting_data
```

### 4. Create a boolean column 'Fast Ship' where 'Shipping Days' < 3

```python
sales_forecasting_data['Fast Ship'] = sales_forecasting_data['Shipping Days'] < 3

sales_forecasting_data.loc[sales_forecasting_data['Shipping Days']< 3]
```

### 5. Replace 'Consumer', 'Corporate', 'Home Office' with short codes ('C', 'CO', 'H')

```python
sales_forecasting_data['Segment'] = sales_forecasting_data['Segment'].replace({'Consumer': 'C', 'Corporate':'CO', 'Home Office': 'H'})

sales_forecasting_data
```

### 6. Capitalize all 'City' names

```python



sales_forecasting_data.City = (sales_forecasting_data.City).str.capitalize()

sales_forecasting_data
```

### 7. Lowercase all product names

```python
sales_forecasting_data['Product Name'] = sales_forecasting_data['Product Name'].str.lower()

sales_forecasting_data
```

### 8. Extact the first 5 characters of 'Product ID'


```python
sales_forecasting_data['Product_ID_Prefix'] = sales_forecasting_data['Product ID'].str[:5]
sales_forecasting_data
```

### 9. Add a new column 'Order Month' using Month name of Order date

```python
sales_forecasting_data['Order Month'] = pd.to_datetime(sales_forecasting_data['Order Date'],dayfirst=True).dt.month_name()

sales_forecasting_data
```

### 10. Replace 'West' with 'W' and vice versa for all values in 'Region' column

```python
sales_forecasting_data['Region'] = sales_forecasting_data['Region'].replace({'West': 'W', 'East': 'E', 'North':'N', 'South': 'S'})

sales_forecasting_data
```
