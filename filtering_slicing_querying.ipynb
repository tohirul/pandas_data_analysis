```python
import pandas as pd
```

```python
sales_forecasting_data = pd.read_csv("./sales-forecasting/train.csv")

sales_forecasting_data.head(20)
```

### 1. Select only 'Order Date', 'Ship Date' and 'Sales' columns

```python
sales_forecasting_data[["Order Date", "Ship Date", "Sales"]].head(5)
```

### 2. Show only the first 5 rows where the category is 'Furniture'

```python
furniture_forecastng_data = sales_forecasting_data.query("Category=='Furniture'").reset_index(drop=True)
furniture_forecastng_data.index+=1

(furniture_forecastng_data).head(5)
```

### 3. Display orders where "Sales" > 500

```python
filtered_sales_data = sales_forecasting_data.query("Sales>500")

filtered_sales_data.sort_values(["Sales"], ascending=True).head(15)
```

### 4. Filter rows where "Region" is "South" and "Sales" > 100

```python
filtered_sales_data = sales_forecasting_data.query("Region=='South' & Sales > 100")
filtered_sales_data.sort_values(["Sales"], ascending=True)
```

### 5. Get all orders with "Ship Mode" equal to "Second Class" or "First Class"

```python

filtered_sales_data = sales_forecasting_data.loc[
    sales_forecasting_data["Ship Mode"].isin(["First Class","Second Class"]) 
    ]
```

```python
filtered_sales_data.sort_values(["Ship Mode"], ascending=True).head(5)
```

```python
filtered_sales_data.sort_values(["Ship Mode"], ascending=False).head(5)
```

### 6. Show orders not belonging to 'Consumer' segment

```python
sales_forecasting_data.loc[sales_forecasting_data["Segment"] != "Consumer"].head(5)
```

### 7. Display orders placed in 'California' State

```python
sales_forecasting_data.loc[sales_forecasting_data['State'] == 'California']
```

### 8. Display all ordres for the city 'Los Angeles'

```python
sales_forecasting_data.loc[sales_forecasting_data['City'] == 'Los Angeles']
```

### 9. Select orders shipped between "2017-01-01" and "2017-12-31"

```python
sales_forecasting_data["Order Date"] = pd.to_datetime(sales_forecasting_data["Order Date"], dayfirst=True)

sales_forecasting_data.loc[sales_forecasting_data["Order Date"].between(pd.to_datetime("2017-01-01",dayfirst=False), pd.to_datetime("2017-12-31", dayfirst=False))].sort_values("Order Date", ascending=True)
```

### 10. Show top 10 highest in 'Sales'

```python
sales_forecasting_data.sort_values("Sales", ascending=False).head(10)
```

### 11. Find order with 'Sales' between 200 and 500

```python
sales_forecasting_data.loc[sales_forecasting_data['Sales'].between(200,500)].sort_values(['Sales', 'Order Date'], ascending=[True, True])
```
