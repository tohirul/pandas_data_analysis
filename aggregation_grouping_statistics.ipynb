### 14. Find which category has the highest average sales

```python
(sales_forecasting_data
    .groupby('Category')['Sales']
    .agg(Total_Count='count', Total_Sales="sum", Average_Sales= 'mean')).round(2).nlargest(1,'Average_Sales')
```

### 15. For each state find the most common 'Ship Mode'

```python
state_common_ship_mode = (
    sales_forecasting_data
    .groupby(['State', 'Ship Mode'])['Ship Mode']
    .agg(Total_Orders='count')
    .reset_index()
    .sort_values(['State', 'Total_Orders'], ascending=[True, False])
    .drop_duplicates(subset=["State"],keep="first")
    .reset_index(drop=True)
    
)

state_common_ship_mode.index+=1

state_common_ship_mode
```

### 16. Find total sales per 'Category', sorted descending

```python
categorized_total_sales = (
    sales_forecasting_data
    .groupby(['Category'])['Sales']
    .agg(Sales_Count='count', Sales_Total='sum')
    .sort_values('Sales_Total',ascending=False)
    .round(2)
)

categorized_total_sales
```

### 17. Compute overall sales per 'Region' per year

```python
regional_overall_sales = (
    sales_forecasting_data.groupby(['Region', pd.to_datetime(sales_forecasting_data['Order Date'],dayfirst=True).dt.year])['Sales']
    .agg(Sales_Count='count', Sales_Total='sum').round(2)
)

regional_overall_sales
```

### 18. For each 'Segment', compute the number of unique products sold

```python
unique_products_sales_forecasting = (
    sales_forecasting_data
    .groupby(['Segment', 'Product ID', 'Product Name'])['Product ID']
    .agg(Total_Count='count')
)

unique_products_sales_forecasting
```

### 19. For each 'Sub-Category', compute the total and mean 'Sales'

```python
sub_catagory_sales_forecasting = (
    sales_forecasting_data
        .groupby('Sub-Category')['Sales']
        .agg(Sales_Count='count', Sales_Total='sum', Sales_Average='mean')
            .sort_values(['Sales_Count', 'Sales_Total', 'Sales_Average'], ascending=[True,True,True])
        .round(2)
)

sub_catagory_sales_forecasting
```

### 20. Display top 3 sub-categories with highest total sales

```python
highest_total_sales = (
    sales_forecasting_data
        .groupby(['Sub-Category'])['Sales']
        .agg(Sales_Count='count', Sales_Total='sum')
        .reset_index()
        .sort_values(['Sales_Total', 'Sales_Count'], ascending=[False, False])
        .round(2)
        .reset_index(drop=True)
)
highest_total_sales.index+=1

highest_total_sales.head(3)
```

```python

def count_salary_categories(accounts: pd.DataFrame) -> pd.DataFrame:
    all_categories = ["Low Salary", "Average Salary", "High Salary"]
    accounts['category'] = (accounts['income']
        .apply(lambda x: 
            'Low Salary' if x < 20000 
             else ('High Salary' 
                    if x > 50000 
                    else 'Average Salary')))
    return pd.DataFrame(
        accounts.groupby(['category', ])
        .size()
        .reindex(all_categories, fill_value=0)
        .reset_index(name="accounts_count"))
```

```python
accounts_category=count_salary_categories(employee_accounts)

(accounts_category)
```
