```python
import pandas as pd
```

```python
data = [
    
    ["Alice",25],["Bob",30],["Charlie",28]
]
list_df =pd.DataFrame(data, columns=["Name","Age"],index=[1,2,3])
print(list_df)
print(type(list_df))
```

```python
data_tuple = (
   ("Alice",25),("Bob",30),("Charlie",28)
)
print(data_tuple)

tuple_df = pd.DataFrame(data_tuple,columns=["Name","Age"],index=[1,2,3])
print(tuple_df)
```

```python
data_dict = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25,30,28]
}
print(data_dict)

dict_df = pd.DataFrame(data_dict,index=[1,2,3])
print(dict_df)
```
